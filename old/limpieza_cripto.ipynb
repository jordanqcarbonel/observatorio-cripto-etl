## Pipeline de Limpieza y Optimización de Datos
Este notebook procesa los datos de criptomonedas para eliminar el ruido aleatorio y guardarlos en formato Parquet.

```python
import pandas as pd
import os
print("✅ Librerías cargadas correctamente")
```

```python
archivo_sucio = "cripto_precios_SUCIO.csv"
df = pd.read_csv(archivo_sucio, dtype=str, encoding="utf-8-sig")

# Limpieza
regex_limpieza = r'[^0-9.]'
columnas_a_limpiar = ['precio_usd', 'cambio_24h', 'volumen_24h', 'cap_mercado']

for col in columnas_a_limpiar:
    if col in df.columns:
        # La lógica exacta de ensuciar_precios.py
        df[col] = (df[col]
                   .str.replace(',', '', regex=False)
                   .str.replace(regex_limpieza, '', regex=True)
                   .str.strip())
        df[col] = df[col].replace('', '0')
        df[col] = pd.to_numeric(df[col], errors='coerce')

print("✅ Limpieza y transformación completada.")
df.head() # Esto reemplaza los print finales
```

```python
df.to_parquet("cripto_precios_OPTIMIZADO.parquet", index=False)
print("🚀 Archivo Parquet generado con éxito. ¡Datos listos para ser analizados!")
```
